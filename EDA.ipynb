```python
# **This command imports the ARIMA model for forecasting**
```

```python
from statsmodels.tsa.arima.model import ARIMA
```

```python
# **This command selects the Local Route data for forecasting**
```

```python
local_series = df['Local Route']
```

```python
# **This command trains the ARIMA model on Local Route data**
```

```python
model = ARIMA(local_series, order=(1,1,1))
model_fit = model.fit()
```

```python
# **This command predicts the next 7 days for Local Route**
```

```python
forecast_local = model_fit.forecast(steps=7)
forecast_local
```

```python
plt.figure(figsize=(10,5))
plt.plot(local_series, label='Original Data')
plt.plot(forecast_local, label='7-Day Forecast', color='red')
plt.legend()
plt.title("Local Route - 7 Day Forecast")
plt.show()
```

```python
light_series = df['Light Rail']

model = ARIMA(light_series, order=(1,1,1))
model_fit = model.fit()

forecast_light = model_fit.forecast(steps=7)
forecast_light
```

```python
last_7 = df.tail(7)
future_dates = pd.date_range(start=df.index[-1] + pd.Timedelta(days=1), periods=7)

forecast = pd.DataFrame({'Date': future_dates})

for col in df.columns:
    if col != 'Day':
        forecast[col] = last_7[col].values

forecast
```
